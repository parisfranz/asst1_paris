```python
speed = 386.550/np.array([386.550, 197.007, 236.134,159.648,156.110,119.908,114.173,96.646])
threads = np.arange(1,len(speed)+1,1)

plt.plot(threads, speed, 'o-')
plt.ylabel('Speedup')
plt.xlabel('Number of Threads')
plt.grid()
```

```python
speed2 = 226.108/np.array([226.108, 134.908, 103.255, 87.253, 75.702, 66.673, 59.134, 49.896])
threads2 = np.arange(1,len(speed2)+1,1)

plt.plot(threads2, speed2, 'o-')
plt.ylabel('Speedup')
plt.xlabel('Number of Threads')
plt.grid()
plt.title('View 2')
```

```python
fig,axs = plt.subplots(nrows = 1, ncols = 2, figsize = (9,4))
plt.subplots_adjust(wspace = 0.3)
axs[0].plot(threads, speed, 'o-')
axs[1].plot(threads2, speed2, 'o-')
axs[0].set_ylabel('Speedup')
axs[0].set_xlabel('Number of Threads')
axs[0].grid()
axs[1].set_ylabel('Speedup')
axs[1].set_xlabel('Number of Threads')
axs[1].grid()
axs[1].set_title('View 2, Serial Time: %1.1f ms' %226.108)
axs[0].set_title('View 1, Serial Time: %1.1f ms' %386.550)
```
